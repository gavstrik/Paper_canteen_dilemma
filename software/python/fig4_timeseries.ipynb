```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
sns.set_context('notebook')
plt.rcParams["font.family"] = "sans-serif"
PLOTS_DIR = '../plots'
```

```python
datafiles =  ['../data/MTurk_anonymous.xlsx', '../data/DTU1_anonymous.xlsx', '../data/DTU2_anonymous.xlsx']
```

```python
df_all = pd.DataFrame()
for datafile in datafiles:
    df = pd.DataFrame(pd.read_excel(datafile))
    df_all = df_all.append(df, sort=True)
df_all = df_all[['round','arrival', 'choice']]
df_all.head()
```

get the total number of choices for each arrival time each round:

```python
df_tot = df_all.groupby(['arrival', 'round'])['choice'].size().unstack(level=0).fillna(0)
df_tot.head()
```

and also get the number of canteen/office choices for each arrival time and round:

```python
df1 = df_all.groupby(['arrival', 'round'])['choice'].value_counts().unstack(level=0).fillna(0)
df1.head()
```

select only the canteen choices:

```python
dfc = df1.iloc[df1.index.get_level_values(1) == 1.0].droplevel(level=1)  # these are the canteen choices
dfc.head()
```

Now we get the frequency of going to the canteen for each arrival time/round by simply dividing dfc with df_tot:

```python
df = dfc/df_tot
df.head()
```

```python
x = np.arange(float(len(dfc)))
colors = ['#00F800', '#2AF598', '#2AF598', '#1BD7BB', '#14C9CB', '#0FBED8', '#08B3E5', '#188AF0']
colors2 = ['#FF7F7F', '#FE0002', '#D80027', '#A1015D', '#63009E', '#2A00D5', '#0302FC', '#00008b']
fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, df[8], s=df_tot[8].values, marker='o', color=colors2[0], label='8:00')
ax.scatter(x, df[8.1], s=df_tot[8.1].values, marker='o', color=colors2[1], label='8:10')
ax.scatter(x, df[8.2], s=df_tot[8.2].values, marker='o', color=colors2[2], label='8:20')
ax.scatter(x, df[8.3], s=df_tot[8.3].values, marker='o', color=colors2[3], label='8:30')
ax.scatter(x, df[8.4], s=df_tot[8.4].values, marker='o', color=colors2[4], label='8:40')
ax.scatter(x, df[8.5], s=df_tot[8.5].values, marker='o', color=colors2[5], label='8:50')
ax.scatter(x, df[9], s=df_tot[9].values, marker='o', color=colors2[6], label='9:00')
ax.scatter(x, df[9.1], s=df_tot[9.1].values, marker='o', color=colors2[7], label='9:10')
ax.plot(x+1, np.poly1d(np.polyfit(x+1, df[8.5].values, 1,
        w=np.sqrt(df_tot[8.5].values)))(x+1), color=colors2[5], label='WLS')
ax.plot(x+1, np.poly1d(np.polyfit(x+1, df[8.4].values, 1,
        w=np.sqrt(df_tot[8.4].values)))(x+1), color=colors2[4])
ax.plot(x+1, np.poly1d(np.polyfit(x+1, df[9].values, 1,
        w=np.sqrt(df_tot[9].values)))(x+1), color=colors2[6])

plt.ylabel('frequency of canteen choice', fontsize=18)
plt.xlabel('round', fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
handles, _ = ax.get_legend_handles_labels()
labels = ['8:00', '8:10','8:20','8:30','8:40','8:50','9:00','9:10', 'WLS']

# bruger np.roll til at placere WLS nederst:
lgd = ax.legend(np.roll(handles,8), labels, title='arrival time', fontsize=14, prop={'size': 14}, bbox_to_anchor=(1, 1))
lgd.legendHandles[8].set_color('grey')
plt.tight_layout()

# Remember: save as pdf and transparent=True for Adobe Illustrator
if not os.path.exists(PLOTS_DIR):
    os.makedirs(PLOTS_DIR)

plt.savefig(os.path.join(PLOTS_DIR, 'fig4_timeseries.png'), bbox_extra_artists=(lgd,), bbox_inches='tight', transparent=True, dpi=300)
plt.savefig(os.path.join(PLOTS_DIR, 'fig4_timeseries.pdf'), bbox_extra_artists=(lgd,), bbox_inches='tight', transparent=True, dpi=300)
sns.despine()
```
